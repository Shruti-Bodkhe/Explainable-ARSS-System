```python
import pandas as pd

df = pd.read_csv("../dataset/resume_cleaned_dataset.csv")
skills_df = pd.read_excel("../dataset/skills_master.xlsx")

print("Resume Dataset Loaded:", df.shape)
print("Skills Dataset Loaded:", skills_df.shape)
```

```python
# Extract skill names column and convert to lowercase list
skill_list = skills_df["skill_name"].str.lower().tolist()

# Print total number of skills extracted
print("Total Skills:", len(skill_list))

# Print first 10 skills to verify extraction
print(skill_list[:10])
```

```python
# Function to extract skills safely

def extract_skills(resume_text):

    detected_skills = []

    # Ensure resume_text is string
    resume_text = str(resume_text).lower()

    for skill in skill_list:
        if skill in resume_text:
            detected_skills.append(skill)

    return detected_skills
```

```python
# Standardize column names for consistency
skills_df.columns = skills_df.columns.str.strip().str.lower()
df.columns = df.columns.str.strip().str.lower()
```

```python
# Remove missing skill names
skills_df = skills_df.dropna(subset=["skill_name"])

# Convert skills into lowercase string list
skill_list = skills_df["skill_name"].astype(str).str.lower().tolist()

print("Total Skills Loaded:", len(skill_list))
```

```python
# Replace missing cleaned_text values with empty string
df["cleaned_text"] = df["cleaned_text"].fillna("")
```

```python
# Function to detect skills from resume text
def extract_skills(resume_text):
    
    detected_skills = []   # Store matched skills
    
    # Safety check → ensure resume_text is string
    if not isinstance(resume_text, str):
        return []
    
    # Check each skill inside resume text
    for skill in skill_list:
        if skill in resume_text:
            detected_skills.append(skill)
    
    return list(set(detected_skills))   # Remove duplicates
```

```python
# Apply skill extraction on all resumes
df["detected_skills"] = df["cleaned_text"].apply(extract_skills)

# View sample results
df[["resume_file", "detected_skills"]].head()
```

```python
# Save dataset with detected skills
df.to_csv("../dataset/resume_with_skills.csv", index=False)

print("Skill extraction completed and dataset saved.")
```
