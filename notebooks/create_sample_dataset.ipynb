```python
import os
import random
import shutil
```

```python
source_folder = "../dataset/resumes_all"
destination_folder = "../dataset/resumes_sample"
```

```python
all_files = os.listdir(source_folder)

print("Total resumes available:", len(all_files))
```

```python
sample_size = 300

sample_files = random.sample(all_files, sample_size)

print("Sample selected:", len(sample_files))
```

```python
for file in sample_files:
    shutil.copy(
        os.path.join(source_folder, file),
        os.path.join(destination_folder, file)
    )

print("Sample dataset created successfully!")
```
