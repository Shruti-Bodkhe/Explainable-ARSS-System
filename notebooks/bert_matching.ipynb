```python
# BERT embedding library
from sentence_transformers import SentenceTransformer

# Similarity calculation
from sklearn.metrics.pairwise import cosine_similarity

# Data handling
import pandas as pd
```

```python
# Load cleaned resume dataset
df = pd.read_csv("../dataset/resume_cleaned_dataset.csv")

print("Dataset Loaded:", df.shape)
df.head()
```

```python
# Load pretrained BERT model
bert_model = SentenceTransformer("all-MiniLM-L6-v2")

print("BERT Model Loaded Successfully")
```

```python
# Sample Job Description
job_description = """
Looking for a Data Scientist with strong skills in Python,
Machine Learning, Deep Learning, SQL and Data Visualization.
"""

print(job_description)
```

```python
# Convert job description into vector embedding
jd_embedding = bert_model.encode(job_description)

print("Job Description Embedding Created")
```

```python
# Convert all resumes into embeddings
resume_embeddings = bert_model.encode(df["cleaned_text"].tolist())

print("Resume Embeddings Created")
```

```python
# Calculate similarity between JD and resumes
similarity_scores = cosine_similarity([jd_embedding], resume_embeddings)[0]

# Store scores
df["bert_match_score"] = similarity_scores

df.head()
```

```python
# Sort candidates by score
ranked_df = df.sort_values(by="bert_match_score", ascending=False)

ranked_df[["resume_file", "bert_match_score"]].head(10)
```

```python
# Save BERT scores
df.to_csv("../results/bert_scores.csv", index=False)

print("BERT scores saved successfully!")
```
