```python
import pandas as pd

df = pd.read_csv("../dataset/job_description.csv")
df.head()
```

```python
def extract_skills(text):
    # Convert to lowercase
    text = text.lower()
    
    # Split by comma
    skills = text.split(",")
    
    # Clean spaces
    skills = [skill.strip() for skill in skills]
    
    return skills

df["job_skills"] = df["job_description"].apply(extract_skills)

df.head()
```

```python
df.to_csv("../dataset/job_with_skills.csv", index=False)

print("Job skill extraction completed.")
```
